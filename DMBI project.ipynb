# 1. Here we have found the number of a particular type of booking i.e ASAP, Reserved, others etc. using these values comapnies can work on the parameters to improve the performance and can do better for their passengers.

```python
count=0
count1=0
count2=0
for i in t1.start_type:
    if(i=="asap"):
        count+=1
    elif(i=="reserved"):
        count1+=1
    else:
        count2+=1
print("asap: ",count)
print("reserved: ",count1)
print("delayed: ",count2)
labels = 'ASAP', 'Reserved', 'Delayed'
sizes = [count,count1,count2]
colors = ['gold', 'yellowgreen', 'blue']
explode = (1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90, radius=1000)
fig = plt.gcf()
fig.set_size_inches(8,8)
plt.legend()
plt.axis('equal')
plt.show()
```

# 2. Here we have calculated the average rating of drivers of the firm. according to the avg. rating, company can decide to strive for better ratings  

```python
score=0
num=0
list10=[]
list11=[]
rate=[]
for i in t2.driver_score:
    list11.append(i)
    score+=i
    num+=1
print(num)
avg=score/(num)
print(avg)
print(len(list11))
for i in list11:
    list10.append(list11.count(i))
    rate.append(i//1)
    while i in list11:
        list11.remove(i)
print(list10)
print(rate)
plt.figure(figsize=(10,10))
plt.bar(rate, list10)
plt.xlabel("Counts",color="blue",size=20)
plt.ylabel("Ratings",color="blue",size=20)
plt.show()
```
